```python
import pandas as pd
pd.set_option("display.max_rows", 20)
pd.set_option("display.max_columns", None)
import numpy as np
import dask.dataframe as dd
```

```python
df_dept = dd.read_csv("departments.csv")
df_dept.head()
```

```python
print(df_dept.shape[0].compute())
```

```python
df_aisles = dd.read_csv("aisles.csv")
df_aisles.head()
```

```python
print(df_aisles.shape[0].compute())
```

```python
df_products = dd.read_csv("products.csv")
df_products.head()
```

```python
print(df_products.shape[0].compute())
```

```python
df_product_details = (
    df_products
    .merge(df_aisles, on='aisle_id', how='inner')
    .merge(df_dept, on='department_id', how='inner')
)
```

```python
print(df_product_details.shape[0].compute())
```

```python
df_product_details.head()
```

```python
import dask.dataframe as dd
```

```python
df_product_details = df_product_details.drop(columns=['aisle_id', 'department_id'])
```

```python
df_product_details = df_product_details.drop_duplicates()
df_product_details = df_product_details.dropna()
```

```python
print(df_product_details.shape[0].compute())
```

```python
df_order_products_prior = dd.read_csv("order_products__prior.csv")
```

```python
df_order_products_train = dd.read_csv("order_products__train.csv")
```

```python
df_order_products = dd.concat([df_order_products_prior, df_order_products_train]).drop_duplicates()
df_order_products.head()
```

```python
print(df_order_products.shape[0].compute())
```

```python
df_orders = dd.read_csv("orders.csv")
df_orders.head()
```

```python
print(df_orders.shape[0].compute())
```

```python
df_order_details = (
    df_orders
    .merge(df_order_products, on='order_id', how='inner')
)
df_order_details.head()
```

```python
print(df_order_details.shape[0].compute())
```

```python
df_order_product_details = (
    df_order_details
    .merge(df_product_details, on='product_id', how='inner')
)
df_order_product_details.head()
```

```python
print(df_order_product_details.shape[0].compute())
```
